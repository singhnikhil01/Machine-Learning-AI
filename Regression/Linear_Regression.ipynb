#**Linear Regression**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("/content/Salary_Data.csv")
data.head()
```

###Selecting columns in the dataset

```python
X_columns =data.iloc[: ,:-1].values
print(X_columns)
```

```python
y_columns = data.iloc[:,-1].values 
print(y_columns)
```

###Splitting Data into Training and Testing Set  

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train, y_test = train_test_split(X_columns,y_columns,test_size = 0.2,random_state=2)
```

```python
print(x_train)
```

```python
print(y_train)
```

###Feature Scaling : Appling feature scaling only after the Spliting the data, 

###**Training a Simple Linear Regression model in a Training set** **bold text**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
predict = reg.predict(x_test)
```

##Visualize the Model

####Visualise the training set

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,reg.predict(x_train),color ='blue')
plt.title("Salary vs Exprience")
plt.xlabel('Exprience')
plt.ylabel('Salary')
plt.show()
```

####Visualise the testing set

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,reg.predict(x_test),color ='blue')
plt.title("Salary vs Exprience")
plt.xlabel('Exprience')
plt.ylabel('Salary')
plt.show()
```
