```python
import numpy as np #work with arrays and math things
import pandas as pd # working on dataset 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
```

##**Loading the dataset** 

```python
data = pd.read_csv('Data.csv')
data.head(10)
```

##**Data Preprocessing** 

```python
#feature and Dependent variable 
#features are the independent variable and labels are the dependent variable 
x = data.iloc[:,:-1] #spliting the dataset # -1 is the index of last column 
#it include the lower bound and exclude the upper bound 
y = data.iloc[:,-1].values # : is the range  in python 
```

###**Taking care of missing data**

```python
##replace the missing data(average , median and the most frequent can be used )
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy="mean") # replace the Nan-values to mean 
# x = imputer.fit(x[:,1:3])
x.iloc[:,1:3] = imputer.fit_transform(x.iloc[:,1:3])
x = x.values
```

```python
type(x)
```

##**Econding Categorical data**

```python
#chaning the String data in to the numerical data 
#Encodinga categorical data  
#onehot encoding - binary vector for each of the country 

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
#mix these two class to transform our data 

ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])] ,remainder='passthrough' ) 
# X = np.array(ct.fit_transform(x[0]))
x = ct.fit_transform(x) #x is already a numpy array so , np.array() is not required 
```

```python
x
```

```python
#when two classes label Encoding , if more classes do onehot Encoding 
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y= le.fit_transform(y)
```

```python
y
```

##**Training testing and spliting**

```python
from sklearn.model_selection import train_test_split 

x_train, x_test ,y_train ,y_test = train_test_split(x, y, test_size=0.2 ,random_state= 1)
#if random state is zero we can get same test and train data 
```

##**Standardization**

```python
#feature scaling is applied on only on the training set 
#don't apply the feature scaling on the dummy variable , that is the encoded variable 
from sklearn.preprocessing import StandardScaler 

sc = StandardScaler()

# Fit and transform the training data
x_train[:,3:] = sc.fit_transform(x_train[:,3:])

# Transform the test data using the same scaler object
x_test[:,3:] = sc.transform(x_test[:,3:])

#fit provides all the necessary information and transform will apply the trnsform on the dataset 
```
