##**Hierarichical Clustering**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('/content/Mall_Customers.csv')
data.head()
```

```python
x = data.iloc[:,[3,4]].values
```

##**Using dendagram to Find the Optimal number of Cluster**



```python

import scipy.cluster.hierarchy as sch 
#method = 'ward' - method of minimum varience 
dendrogram = sch.dendrogram(sch.linkage(x,method='ward') )
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

##**Hierarichal Clustering Model** 

```python
from sklearn.cluster import AgglomerativeClustering
#affinity is type of distance 
cluster = AgglomerativeClustering(n_clusters= 5 , affinity='euclidean',linkage='ward')
y_hc = cluster.fit_predict(x)
```

```python
print(y_hc)
```

##**Visualizing the Clustering** 

```python
plt.scatter(x[y_hc == 0,0],x[y_hc == 0,1],s=20,color='red',label='cluster1')
plt.scatter(x[y_hc == 1,0],x[y_hc == 1,1],s=20,color='blue',label='cluster2')
plt.scatter(x[y_hc == 2,0],x[y_hc == 2,1],s=20,color='green',label='cluster3')
plt.scatter(x[y_hc == 3,0],x[y_hc == 3,1],s=20,color='cyan',label='cluster4')
plt.scatter(x[y_hc == 4,0],x[y_hc == 4,1],s=20,color='magenta',label='cluster5')
plt.title('clusters of customers')
plt.ylabel('Spending Score (1-100)')
plt.xlabel('Annual Income (k$)')
plt.legend()
plt.show()
```
