```python
import networkx as nx 
import matplotlib.pyplot as plt 
```

```python
edges =[(1,2,5),(2,3,10),(2,4,15),(4,5,20)]
```

```python
G = nx.Graph()
```

```python
G.add_weighted_edges_from(edges)
```

```python
nx.draw(G)
```

```python
pos = nx.circular_layout(G)
nx.draw(G,pos=pos)
```

```python
nx.draw_networkx(G,pos, with_labels=True,node_color='Green' )
```

```python
nx.draw(G,pos,with_labels=True,node_color='Green')
```

```python
nx.draw_networkx_edge_labels(G,pos=pos)
```

#Charachers in Graphs 

```python
g = nx.Graph()
```

```python
g.add_edges_from([
      ("A","B",{"W":5}),
      ("A","C",{"W":6}),
      ("C","D",{"W":7}),
      ("D","B",{"W":7}),
      ("B","D",{"W":9})
                                    ])
```

```python
pos = nx.circular_layout(g)
```

```python
labels = nx.get_edge_attributes(g,'W')
```

```python
nx.draw(g,pos, with_labels=True)
```

```python
nx.draw_networkx_edge_labels(g,pos=pos)
```
