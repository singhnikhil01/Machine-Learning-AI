#**Random Forest Regressor**  (Ensemble Learning )

###**Feature scaling must not be applied to the Decision Trees & Random Forest Classifier**


**Ensemble learning** is the process by which multiple models, such as classifiers or experts, are strategically generated and combined to solve a particular computational intelligence problem

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('/content/Position_Salaries.csv')
```

```python
data.head()
```

```python
x = data.iloc[:,1:2].values
y = data.iloc[: ,-1].values
```

```python
x 
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor 
regressor = RandomForestRegressor(n_estimators= 10,min_samples_split= 2,random_state= 3)
regressor.fit(x,y) 
```

```python
regressor.predict(x)
```

```python
regressor.predict([[6.5]])
```

```python
x_grid = np.arange(min(x),max(x),0.001)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x,y)
plt.plot(x_grid,regressor.predict(x_grid),color='red')
plt.grid()
plt.show()
```

#**R-Squared**(Goodness of Fit)

###**The way we predict is to minimize the mean squared error**

SS(res) = sum(yi-yi(pred))^2

SS(tot) = sum(yi - yi(avg))^2

R^2 =  1- SS(res)/SS(tot) [higher value is the good fit]

if model is having downward slope i.e line up to down so, that time ss(tot) is preferred 

**K - number of dependent variable** 

**n - sample size**
