#**How to select the regression  Model?**(based on Dataset)

```python
import numpy as np 
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
data = pd.read_csv('Data.csv')
```

```python
data.head()
```

```python
x = data.iloc[:,:-1].values
x 
```

```python
y = data.iloc[:,-1].values
y
```

##**Split Train and Test**

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y ,test_size = 0.2,random_state=0)
```

##**Multiple Regression**

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
ypred_lin = lin_reg.predict(x_test)
print(ypred_lin)
```

```python
print(r2_score(y_test, ypred_lin))
```

##**Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
poly.fit(x_train)
xpoly = poly.transform(x_train)
xpoly
```

```python
from sklearn.linear_model import LinearRegression
lin_regpoly = LinearRegression()
lin_regpoly.fit(xpoly,y_train)
```

```python
ypred_linpoly = lin_regpoly.predict(poly.transform(x_test))
print(ypred_linpoly)
```

```python
print(r2_score(y_test, ypred_linpoly))
```

##**SVM Regression**

```python
from sklearn.preprocessing  import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
sc_x.fit(x_train)
sc_y.fit(y_train.reshape(-1,1))
```

```python
sc_y.transform(y_train.reshape(-1,1))
```

```python
y_train.reshape(1,len(y_train))
```

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf')
svr_reg.fit(sc_x.transform(x_train),sc_y.transform(y_train.reshape(-1,1)))
```

```python
ypred_svr = svr_reg.predict(sc_x.transform(x_test))
ypred_svr = sc_y.inverse_transform(ypred_svr.reshape(-1,1))
print(ypred_svr)
```

```python
print(r2_score(y_test, ypred_svr))
```

##**Desision Tree Regression**

```python
from sklearn.tree import DecisionTreeRegressor 
des_reg = DecisionTreeRegressor(random_state = 2 )
des_reg.fit(x_train,y_train)
```

```python
ypred_des = des_reg.predict(x_test)
print(ypred_des)
```

```python
print(r2_score(y_test, ypred_des))
```

##**Random Foerest Regression**

```python
from sklearn.ensemble import RandomForestRegressor 
ran_reg = RandomForestRegressor(n_estimators= 10,random_state = 2 )
ran_reg.fit(x_train,y_train)
```

```python
ypred_ran = ran_reg.predict(x_test)
print(ypred_ran)
```

```python
print(r2_score(y_test, ypred_ran))
```
