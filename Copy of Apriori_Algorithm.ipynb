##**Change the data Format**

```python
#change the format so that data can be fit in the apriori model 
transactions =[]

#iterating through each row can column to change the format from data frame to list 
for i in range(0, 7501):
  transactions.append([str(data.values[i,j]) for j in range(0, 20)])
```

##**Training Apriori model on the dataset our Model** 

```python
from apyori import apriori
#minimum number we want see per day , 3*7/75001 (75001total data,min 3 transaction per day )
# 7 days per week , 3 is good lift 
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

#**Visualizing the Result** 

##**Displaying The first Result from the output of apriori function** 

```python
results = list(rules)
results 
```

##**Putting the Results well Organised into Pandas Data Frame** 

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    # confidences = [result[2][0][2] for result in results]
    # lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```
