##**PART A : Prerequisite for PCA implementation**

image.png

```python
import matplotlib.pyplot as plt 
x1=[11,15,10,18,21,9,34,89,54,67]
x2=[8,18,5,1,16,78,86,97,31,65]
```

```python
import pandas as pd 
d = {'x1': [11,15,10,18,21,9,34,89,54,67], 'x2': [8,18,5,1,16,78,86,97,31,65]}
data = pd.DataFrame(d)
```

```python
data
```

```python
plt.scatter(x1,x2)
```

image.png

```python
import numpy as np 
x_mean = data - np.mean(data,axis=0)
x_mean
```

image.png

```python
cov_mat = np.cov(x_mean,rowvar = False)
```

```python
cov_mat
```

image.png

```python
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
```

```python
eigen_values
```

```python
eigen_vectors
```

image.png

```python
#shorting the index to interchange the eigen values 
sorted_index = np.argsort(eigen_values)[::-1]
sorted_index
```

```python
sorted_eigenvalue = eigen_values[sorted_index]
sorted_eigenvalue
```

```python
sorted_eigenvectors = eigen_vectors[:,sorted_index]
sorted_eigenvectors
```

image.png

```python
PCA_first = sorted_eigenvalue[0]
PCA_first
```

image.png

```python
v1 = sorted_eigenvectors[0]
v1
```

```python
np.dot(np.transpose(v1),np.transpose(x_mean))
```
